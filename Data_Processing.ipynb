```
male_points  = []
female_points = []
data_count_male = len(os.listdir("SPRING_MALE"))
for filename in os.listdir("SPRING_MALE")[:data_count_male]:
  points, _ = trimesh.sample.sample_surface_even(trimesh.load(os.path.join("SPRING_MALE", filename)), 3072)
  female_points.append(points)

data_count_female = len(os.listdir("mesh"))
for filename in os.listdir("mesh")[:data_count_female]:
  points, _ = trimesh.sample.sample_surface_even(trimesh.load(os.path.join("mesh", filename)), 3072)
  male_points.append(points)
```

```
female_points = np.array(female_points)
male_points = np.array(male_points)
print(female_points.shape, male_points.shape)
```
